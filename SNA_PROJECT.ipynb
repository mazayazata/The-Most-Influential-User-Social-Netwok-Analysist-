```
import pandas as pd
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```
#Membaca dataset
df_covid = pd.read_csv("crawling_covid_indonesia.csv")
```

```
#Menampilkan 5 data pertama
df_covid.head()
```

```
#Data preparation kolom yang digunakan hanya kolom user dan user yang dimention
df_covid = df_covid.loc[:,['username','user_mentions']]
df_covid = df_covid.dropna()
df_covid.head()
```

```
#Menghitung jumlah edge
graph_covid = nx.from_pandas_edgelist(df_covid,source="username",target="user_mentions")

nx.info(graph_covid)

graph_covid.nodes()

len(graph_covid.nodes())

len(graph_covid.edges())
```

```
# Visualisasi data
nx.draw
nx.draw_networkx

plt.figure(figsize=(20,20))
nx.draw(graph_covid,with_labels=True)
```

```
# Analisis
# Menghitung derajat dari nodes
tap = nx.degree(graph_covid)
for i in tap:
    print(i[0],' : ',i[1])
    

# Mengidentifikasi Most Influential User dengan menghitung betweeness centrality
nx.degree_centrality(graph_covid)

sorted(nx.betweenness_centrality(graph_covid).values())

most_influential = nx.betweenness_centrality(graph_covid)

for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])

# nx.betweenness_centrality(graph_covid)


plt.show()
```
